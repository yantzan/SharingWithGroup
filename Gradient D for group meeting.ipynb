```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.power(x,2)+x-2.0
```

```python
def df1(x):
    return 2.0*x
```

```python
def df2(f,x,delta = 1e-4):
    return (f(x+delta)-f(x-delta))/(2.0*delta)
```

```python
x_r = np.arange(-6,5,0.01)
plt.plot(x_r,f(x_r))
plt.show()
```

```python
lr = 0.1;#learning rate
x_init = -5
x = x_init

max_iter = 40
plt.plot(x_r,f(x_r))
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
for i in range(max_iter):
    dfx = df2(f,x)
    x = x-lr*dfx
    plt.scatter(x,f(x))
plt.savefig('iterleft.jpg',dpi=300)    
print('initial x=', x_init)
print('argmin f(x) of x =',x)
print('f(x) = ', f(x))
```
